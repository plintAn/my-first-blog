```python
import sqlite3
import pandas as pd

# SQLite 데이터베이스에 연결
sqlite_conn = sqlite3.connect(r'C:\Users\hospc\djangogirls\db.sqlite3')

# 모든 테이블 이름 조회
cursor = sqlite_conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()

# 각 테이블에 대해 데이터 추출 및 분석
for table in tables:
    table_name = table[0]
    print(f"\nAnalyzing table: {table_name}")
    
    # 데이터 추출
    df = pd.read_sql_query(f"SELECT * FROM {table_name}", sqlite_conn)
    
    # 데이터프레임의 기본 정보 출력
    print(f"\nDataFrame Information for {table_name}:")
    print(df.info())
    
    # 데이터프레임의 기본 통계 요약 출력
    print(f"\nStatistical Summary for {table_name}:")
    print(df.describe(include='all'))
    
    # 특정 열에 대한 값의 분포 출력 (예: 'author' 열, 존재하는 경우)
    if 'author' in df.columns:
        print(f"\nValue Counts for 'author' Column in {table_name}:")
        print(df['author'].value_counts())
    elif 'id' in df.columns:
        print(f"\nSample data from 'id' Column in {table_name}:")
        print(df['id'].value_counts().head())

# 연결 종료
sqlite_conn.close()
```
